```python
from  random import randint
df_ptInRange = pd.DataFrame(columns=['point_ID','거리', '밝기', '분류', '명칭', '위도', '경도'])
n = int(0)

for i in range(0, len(total_df)):
    
    # 점과 출발점 거리측정 (strt는 이후에 경유지로 재설정 되어야한다.)
    dis = haversine(strt, [total_df.loc[i,'위도'] , total_df.loc[i,'경도'] ], unit='m')
    
    ## 1 block 내에 포함되는 points
    if dis <= 50  :
        tmp = {'point_ID' : 'Point_%d' % int(n), '분류' :total_df.loc[i,'분류'], '명칭' : total_df.loc[i,'명칭'], '위도' : total_df.loc[i,'위도'], '경도' : total_df.loc[i,'경도'] , '거리' : dis, '밝기' : randint(0,5)  } 
        
        # 아래 코드의 마지막 .T는 transform
        df_ptInRange = df_ptInRange.append(df_ptInRange.from_dict(tmp, orient = 'index').T)
        n += 1

# ascending 오름차순 
a = df_ptInRange.sort_values(by = ['거리','밝기'], ascending=[True, False] ).reset_index(drop=True)
b = df_ptInRange.sort_values(by = ['밝기', '거리'], ascending=[False, True] ).reset_index(drop=True)

# 주변 시설물의 개수라고 생각하면 될듯
print(len(a))

b
```

```python
#1. g 구하기 (출발점으로부터의 거리)
def get_gf(sp, ep, nodes_df ):
    gs = []
    hs = []
    print(type(nodes_df))
    
    for i in range(0, len(nodes_df)) :
        tmp = [ nodes_df.loc[i,'위도'], nodes_df.loc[i,'경도'] ]
        gs.append(haversine(sp,tmp , unit='m'))
        hs.append(haversine(ep, tmp, unit='m'))    
       
    
    return gs, hs
```

```python
#g값, f값 받아오기
g, h = get_gf(strt, dstn, b.loc[:, ['위도','경도']] )

f = []
for i in range(0, len(g)):
    f.append( g[i] + h[i] ) 
    
b['F'] = f
```

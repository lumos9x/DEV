# 밝기 합치기

```python
from  random import randint
df_ptInRange = pd.DataFrame(columns=['point_ID','거리', '밝기', '분류', '명칭', '위도', '경도'])
light = []
for i in range(0, len(total_df)) : 
    light.append(randint(0,5))
```

```python

n = int(0)

for i in range(0, len(total_df)):
    # 점과 출발점 거리측정 (strt는 이후에 경유지로 재설정 되어야한다.)
    dis = haversine(strt, [total_df.loc[i,'위도'] , total_df.loc[i,'경도'] ], unit='m')
    
    ## 1 block 내에 포함되는 points
    if dis <= 80  :
        tmp = {'point_ID' : 'Point_%d' % int(n), '분류' :total_df.loc[i,'분류'], '명칭' : total_df.loc[i,'명칭'], '위도' : total_df.loc[i,'위도'], '경도' : total_df.loc[i,'경도'] , '거리' : dis, '밝기' : light[i]  } 
        
        # 아래 코드의 마지막 .T는 transform
        df_ptInRange = df_ptInRange.append(df_ptInRange.from_dict(tmp, orient = 'index').T)
        n += 1

# ascending 오름차순 
a = df_ptInRange.sort_values(by = ['거리','밝기'], ascending=[True, False] ).reset_index(drop=True)
b = df_ptInRange.sort_values(by = ['밝기', '거리'], ascending=[False, True] ).reset_index(drop=True)

# 주변 시설물의 개수라고 생각하면 될듯
print(len(a))

b
```

```python
# 역삼동 중심 = 역삼역 /  좌표 : 37.500742, 127.036891
to_map1 = folium.Map(location=strt, zoom_start=18)

for n in range(0, len(a)) :
    if a['분류'][n] == '편의점' :
        folium.Marker([a['위도'][n], a['경도'][n]], popup= a['분류'][n], icon=folium.Icon(color='green')).add_to(to_map1)
    elif a['분류'][n] == 'CCTV' : 
        folium.CircleMarker([a['위도'][n], a['경도'][n]], popup= a['point_ID'][n] + a['명칭'][n], radius = a['밝기'][n] + 1).add_to(to_map1)        
    elif a['분류'][n] == '보안등' : 
        folium.CircleMarker([a['위도'][n], a['경도'][n]], popup= a['point_ID'][n] + a['명칭'][n], radius = a['밝기'][n] + 1).add_to(to_map1)  
    else : 
        folium.Marker([a['위도'][n], a['경도'][n]], popup= a['분류'][n], icon=folium.Icon(color='black')).add_to(to_map1)        

        
# 구분하기 쉽게 스타팅 지역은 빨간색 마커로 표시        
folium.Marker(strt, popup= 'SP', icon=folium.Icon(color='red', )).add_to(to_map1)
to_map1
```

```python
#1. g 구하기 (출발점으로부터의 거리)
def get_gf(sp, ep, nodes_df ):
    gs = []
    hs = []
    print(type(nodes_df))
    
    for i in range(0, len(nodes_df)) :
        tmp = [ nodes_df.loc[i,'위도'], nodes_df.loc[i,'경도'] ]
        gs.append(haversine(sp,tmp , unit='m'))
        hs.append(haversine(ep, tmp, unit='m'))    
       
    
    return gs, hs
```

```python
#g값, f값 받아오기
g, h = get_gf(strt, dstn, b.loc[:, ['위도','경도']] )

f = []
for i in range(0, len(g)):
    f.append( g[i] + h[i] ) 

# F = 현재 지점의 출발지에서의 거리 + 목적지에서부터의 거리
b['F'] = f
```

```python
df_ds = b.sort_values(by = ['F']).reset_index( drop= True )
df_ds
```

```python
#최대, 최소.. 정규화?
ds = df_ds.loc[len(b) - 1, 'F'] - df_ds.loc[0,'F']
ls = b.loc[0,'밝기'] - b.loc[len(b) - 1, '밝기'] 

print(ds, ls)
```

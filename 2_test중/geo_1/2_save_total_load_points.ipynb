```python
import json
from pyproj import Proj, transform
import googlemaps
import folium
import pandas as pd
```

```python
load_df = pd.read_csv("all_coords.csv", encoding = 'EUC-KR')
feature_df = pd.read_csv("total.csv", encoding = 'EUC-KR')
```

```python
light_df = feature_df[feature_df.분류 == '보안등']
light_df = light_df.drop(columns = ['Unnamed: 0'])
light_df.head(5)
```

```python
load_df = load_df.loc[:, ['RN', 'LAT', 'LNG']]
load_df['분류'] = '도로'
load_df.head(1)
```

```python
load_df.rename(index=str, columns={ "RN" : "명칭","LAT"  : "위도", "LNG" : "경도" },inplace=True)
load_df.head(4)
```

```python
total_load = pd.concat([load_df,light_df]).reset_index(drop=True)
total_load.head(4)
```

```python
total_load.loc[:,['명칭','분류', '위도','경도' ]].to_csv("total_load_points.csv", mode='w', encoding='ms949') # edge(=link) list
```

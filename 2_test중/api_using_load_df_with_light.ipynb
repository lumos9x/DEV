```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
df = filteredDF.drop(['n','명칭', '분류'],1)
```

```python
model = KMeans(n_clusters=3, algorithm='auto')
model.fit(df)
predict = pd.DataFrame(model.predict(df))
predict.columns = ['predict']

r = pd.concat([df, predict], axis = 1)
r
```

```python
plt.scatter(r['lat'], r['lon'], c=r['predict'], alpha = 5)
plt.show()
```

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.lmplot('lat', 'lon', data = df, fit_reg= False, scatter_kws = {"s":5})

plt.title('kmeans plot')

plt.xlabel('lat')

plt.ylabel('lon')
```

```python
ks = range(1,10)
inertias = []

for k in ks:
    model = KMeans(n_clusters= k)
    model.fit(df)
    inertias.append(model.inertia_)

plt.plot(ks,inertias,'-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```
